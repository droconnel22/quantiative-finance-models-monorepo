### Universe Definition

However, there is a less obvious, yet absolutely critical, reason for paring down the security list: successful cross-sectional strategies balance a tension between price dispersion and self-similarity in the universe. By definition, cross-sectional strategies extract relative value across securities and, in order to be able to rank something intelligently, there needs to be some degree of uniformity in the characteristics of the things being ranked.

### Alpha Combination

Rarely is any single alpah significant enough to be the sole basis of an investment strategy. A successful strategy usually includes many individual alphas; if they are strong enough, a few can suffice.

The goal at this stage is to implmenent a weighting scheme which takes as input many normalized alphas and produces a single alpha which is more predictive than the best individual alpha. 

The weighting scheme can be quite simple: sometimes just adding ranks or averaging youor laphas can be an effective solutioon. 

In fact one popular mode does just that to combine two alphas. For increased complexity classic portfolioo theoory can help you.

For example try solving for the weights such that the final combined alpha has the lowest possible variance.

Lastly modern machine learning techniques can capture complex relationships between alphas.

### Portfolio Construction

We have a final alpha vector and we must introduce this alpha to the real world and structure and trade a portfolioo to capture value.

At each cycle time of the process we calculate the alphas, combine the alphas into a final alpha, inherit the portfolio from the previous period, take our final alpha and calculate a new ideal 

i.e., your longs could be equal weights in names in the top quintile and your shorts could be equal weights in names in the bottom quintile with weights set so the portfolio reaches some total invested value long and short, and such that the value of longs equals the (absolute) value of shorts.
