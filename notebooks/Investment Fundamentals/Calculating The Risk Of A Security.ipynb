# Security's Risk

Your money will earn a stable amount over time
vs
Large variability from year to year, stand to lose 40 of your value

## Variability is the best measure of risk

Investors don't like surprises
- Known rate of return
- Investors are risk adverse
- Reduce it as much as possible

S^2 variance
S standard deviation


S^2 Variance
Measures the dispersion of a set of data points around the mean


S^2 (variance) = sum(data point - mean of data)^2 / sample size (N) - 1


S^2 = sum(x-mean(x))^2 / n-1

S = sqrt(s^2)

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tickers = ['PG','BEI.DE']

sector_data = pd.DataFrame()
```

```python
for ticker in tickers:
    sector_data[ticker] = web.DataReader(ticker,
                                        data_source='yahoo',
                                        start='2007-1-1')["Adj Close"]
```

```python
sector_data.head()
```

```python
sector_data.tail()
```

```python
sector_data.plot(figsize=(16,6));
```

```python
normalize = (sector_data/ sector_data.iloc[0]) * 100
```

```python
normalize.plot(figsize=(16,6));
```

```python
sector_returns = np.log(sector_data/sector_data.shift(1))
```

```python
sector_returns.head()
```

```python
sector_returns['PG'].mean()
```

```python
sector_returns['PG'].std()
```

```python
sector_returns['PG'].std() * 250 ** 0.5
```
