```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
%matplotlib inline
```

```python
# SP500, Nasdaq, German Dax
tickers = ['^GSPC','^IXIC','^GDAXI']

index_data = pd.DataFrame()

for ticker in tickers:
    index_data[ticker] = web.DataReader(ticker,
                                       data_source='yahoo',
                                       start='2000-01-01')['Adj Close']
```

```python
index_data.head()
```

```python
index_data.tail()
```

```python
normalized = ( index_data / index_data.iloc[0]) * 100
```

```python
normalized.head()
```

```python
normalized.plot(figsize=(16,6));
```

```python
# Calculate returns
index_returns = (index_data/index_data.shift(1)) -1
```

```python
index_returns.tail()
```

```python
annual_index_returns = index_returns.mean() * 250
```

```python
annual_index_returns
```
