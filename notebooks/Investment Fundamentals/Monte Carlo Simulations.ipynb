## Corporate Finance Example
1. Revenue Growth 
2. Revenue volatility 
3. Cogs (Cost Of Goods Sold) modeled as a percentage of revenues = Gross Profit
4. Opex (operating expenses) modeled as a percentage of revenues = Operating Profit

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rev_m = 170
rev_stddev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m,rev_stddev,iterations)
rev
```

```python
plt.figure(figsize=(16,6))
plt.plot(rev)
plt.show()
```

```python
COGS = -(rev * np.random.normal(0.6,0.1))
```

```python
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.axhline(COGS.mean(),c='r')
plt.axhline(COGS.mean()+COGS.std(),c='r',ls='--')
plt.axhline(COGS.mean()-COGS.std(),c='r',ls='--')
plt.show();
```
