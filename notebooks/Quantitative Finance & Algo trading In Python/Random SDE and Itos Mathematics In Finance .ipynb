```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as npr
```

```python
def wiener_process(dt=0.1, x0=0,n=1000):
    # W(T=0) = 0
    # initialize W(t) with zeros
    W = np.zeros(n+1)
    
    # we create N+1 timesteps t = 0,1,2,3,...,N
    t = np.linspace(x0,n, n+1)
    
    # we have to use cumulative sum on every step the additional value is 
    # drawn from a normal distribution with mean 0 and variance dt ... N(0,dt)
    
    # note: either express is equivalent
    # by the way: N (0, dt) = sqrt(dt) * N(0,1) usually this formula is used.
    # for dt we take the square root of variance which is the standard deviation
    
    W[1: n+1] = np.cumsum(npr.normal(0,np.sqrt(dt),n))
    
    return t, W
    
    
```

```python
def plot_process(t, W):
    plt.plot(t,W)
    plt.xlabel('Time (t)')
    plt.ylabel('Wiener-process W(t)')
    plt.title('Wiener-Process')
    plt.show();
```

```python
t,W = wiener_process()
plot_process(t,W)
```

```python
t,W = wiener_process(dt=10)
plot_process(t,W)
```

```python
t,W = wiener_process(dt=100)
plot_process(t,W)
```

```python
t,W = wiener_process(dt=1000)
plot_process(t,W)
```

```python
t,W = wiener_process(dt=10000)
plot_process(t,W)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# s0 the initial value of the stock
# T is the end of the simulation
# N sample size
# mu is the mean of the random walk
# sigma random fluctation around the mean

def simulate_geometric_random_walk(s0, T=2, N=1000, mu=0.1, sigma=0.05):
    
    # the incremental time step
    dt = T / N
    
    # sample points
    t = np.linspace(0, T, N)
   
    
    # generate random work or wiener-process
    # standard normally distributed items betweens 0 and 1 for N items
    # mean is 0 and standard devitation is 1 in the default case
    # N(0,dt) = sqrt(dt) * N(0,1)
    W = np.random.standard_normal(size=N)
    
    W = np.cumsum(W) * np.sqrt(dt)
    
    # geometic brownian motion stochastic differential equation
    X = (mu - 0.5 * sigma ** 2) * t + sigma * W
    
    S = s0 * np.exp(X)
    
    #print(f"dt={dt}, t={t} W={W} X={X}")
    
    return t, S
    
    
```

```python

stock_initial_price = 10
t, S= simulate_geometric_random_walk(stock_initial_price)
```

```python
def plot_simulation(t, S):
    plt.plot(t, S)
    plt.xlabel('Time (t)')
    plt.ylabel('Stock Price S(t)')
    plt.title('Geometric Brownian Motion')
    plt.show()
```

```python
plot_simulation(t,S)
```

```python
stock_initial_price = 10
t, S= simulate_geometric_random_walk(stock_initial_price,sigma=0.005)
plot_simulation(t,S)
```

```python
stock_initial_price = 10
t, S= simulate_geometric_random_walk(stock_initial_price,sigma=0.5)
plot_simulation(t,S)
```

```python
stock_initial_price = 55
t, S= simulate_geometric_random_walk(stock_initial_price,mu=0,sigma=2)
plot_simulation(t,S)
```
