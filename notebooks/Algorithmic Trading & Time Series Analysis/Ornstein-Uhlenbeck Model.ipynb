```python
np.random.seed(420)
```

```python
mean = 0
std= 1
```

```python
samples = np.random.normal(mean,std,size=10000)
```

```python
dt=0.1
theta = 1.2
mu =0.5
sigma=0.3
n=10000
```

```python
# X(t=0) = 0 and initalize x(t) with zeros
x = np.zeros(n)
```

```python
# W(t+dt) = - W(t) ~ N(0,dt)
for t in range(1,n):
    x[t] = x[t-1] + theta * (mu - x[t-1])*dt + sigma * np.random.normal(0,np.sqrt(dt))
```

```python
x
```

```python
plt.plot(x)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Ornstein-Uhlenbeck Process')
plt.show();
```
