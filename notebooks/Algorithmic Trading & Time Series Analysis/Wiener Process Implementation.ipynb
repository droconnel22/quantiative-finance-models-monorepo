```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define Geometric Brownian Motion
# https://en.wikipedia.org/wiki/Brownian_motion
# Brownian motion, or pedesis (from Ancient Greek: πήδησις /pɛ̌ːdɛːsis/ "leaping"), is the random motion of particles suspended in a medium (a liquid or a gas).[


# ds = u S dt + o S dW

# where S is the stock price

# u is the mean
# and o is the standard deviation
# dt is change in time
# dW is change in weiner log normal or geometric distribution
```

```python
dt =0.1
x0 = 0
n=1000

# W(t=0) = 0
# initialize W(t) with zeros
W = np.zeros(n+1)

# we create N+1 timesteps: t=0,1,2,3,...,N
t = np.linspace(x0,n,n+1)

# we have to use cumulative sum on every step the additiona value is
# drawn from a normal distribution with mean 0 and variance dt ... N(0,dt)
# by the way N(0, dt) = sqrt(dt) * N(0,1) which is common formula
W[1:n+1] = np.cumsum(np.random.normal(0,np.sqrt(dt),n))

plt.plot(t,W)
plt.xlabel('Time(t)')
plt.ylabel('Winer-process W(t)')
plt.title("Wiener-process")
plt.show();
```

```python

# Initital value of stock
s0 = 10.00
# time periods
T=100
# total number of elements
N=1000
# Mean 
mu = 0.001
# Random flunction around the mean
sigma=0.15

dt = T/N
t = np.linspace(0,T,N)

# standard normal distribution with N(0,1)
# mean zero and std dev of 1
W = np.random.standard_normal(size=N)

# N(0,dt) == sqrt(dt) * N(0,1)
W = np.cumsum(W) * np.sqrt(dt)
X = (mu - 0.5 * sigma ** 2) * t + sigma * W
S = s0 * np.exp(X)

# geometric think log and its inverse transformation function: e ^

plt.plot(t,S)
plt.xlabel('Time(t)')
plt.ylabel('Stock Price S(t)')
plt.title("Geometric Brownian Motion")
plt.show();
```
