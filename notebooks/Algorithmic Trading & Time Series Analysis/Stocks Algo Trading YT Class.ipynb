```python
ret = normalized_close_df.pct_change().dropna()
ret.head()
```

```python
ret.describe().T
```

```python
summary_df = ret.describe().T.loc[:,["mean","std"]]
summary_df
```

```python
summary_df["mean"] = summary_df["mean"] * 252
summary_df["std"] = summary_df["std"] * np.sqrt(252)
summary_df
```

```python
summary_df.plot.scatter(x="std",y ="mean",figsize=(16,10),fontsize=10,s=50)
for i in summary_df.index:
    plt.annotate(i,xy=(summary_df.loc[i,"std"]+0.002, summary_df.loc[i,"mean"]+0.002),size=15)
plt.xlabel('Annual risk(std)',fontsize=10)
plt.ylabel('Annual return',fontsize=10)
plt.title('Risk / Return ',fontsize=14)
plt.show();
```
